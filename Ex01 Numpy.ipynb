### boolean indexing
- 배열 안에서 조건에 만족하는 True인 값들만 추출하는 인덱싱 방법

```python
# 불리언 데이터 만들기 - 데이터를 True / False 형태로 만들어주기
rs = (bmi >= 23) & (bmi <25)
rs
```

```python
# 불리언 인덱싱 진행하기
bmi[rs]
```

- 기존에 인덱싱을 진행하던 방법 : 인덱스 번호를 이용했다.
- 불리언 인덱싱을 진행하는 방법 : 인덱스 번호가 들어가는 곳에 조건식을 걸어, True / False로 이루어진 데이터를 넣어준다
    - 불리언 인덱싱을 진행하면 값이 True인 데이터들만 가지고 온다

- 불리언 인덱싱 이해하기

```python
rd_int = np.random.randint(1, 101, size = (10,10))
```

```python
# 위의 데이터에서 80이상인 데이터만 가져오기
# 불리언 데이터 만들기(True,False)
# 조건 확인 -> 80 이상인지 아닌지 확인하기
rd_int >= 80
```

```python
# 위에서 만든 불리언 데이터를 활용해서 인덱싱을 진행하자.
rd_int[(rd_int % 10 == 8) | (rd_int & 5 == 0) ]
```
